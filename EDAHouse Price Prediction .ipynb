 Exploratory Data Analysis (EDA) Report  House Price Prediction
Introduction

The purpose of this Exploratory Data Analysis (EDA) is to understand the house price dataset and identify the key factors influencing property prices. EDA helps in discovering patterns, relationships, and data issues before building a machine learning model.

Dataset Overview

The dataset contains housing information such as area, number of bedrooms, bathrooms, location, and price. Each row represents one house, and each column represents a feature. The target variable for prediction is Price.

Data Structure

The dataset includes both numerical and categorical variables. Numerical features include area, bedrooms, and bathrooms, while categorical features include location and furnishing status. Correlation analysis was applied only to numerical features.

Data Quality Checks

The dataset was checked for missing values and duplicate records. Some categorical variables were identified that may require encoding before model training. Outliers were observed in the price distribution.

Analysis & Findings

Univariate analysis showed that most house prices fall within a moderate range, with a few high-value outliers.
Bivariate analysis indicated a positive relationship between area and price, meaning larger houses generally cost more. Houses with more bedrooms and bathrooms also tend to have higher prices. Location appears to influence price variation as well.

Correlation analysis revealed that area, bedrooms, and bathrooms have positive correlations with price, making them important predictors.

Conclusion

The dataset is suitable for house price prediction after basic preprocessing. Key factors affecting price include area, number of rooms, and location. Handling outliers and encoding categorical variables will improve model performance. EDA provided valuable insights for the next modeling steps.
